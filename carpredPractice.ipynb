# Apply One Hot Encode on Condition Column And Merged with Data set

```python
def trans_Condition(x):
    if x=='New':
        return 1
    if x=='Used':
        return 2
```

```python
new_df['Tran_Condition']=new_df['Condition'].apply(trans_Condition)
new_df
```

# Apply One Hot Encode on Fuel Column And Merged with Data set

```python
def trans_Fuel(x):
    if x=='CNG':
        return 1
    if x=='Diesel':
        return 2
    if x=='Hybrid':
        return 3
    if x=='LPG':
        return 4
    if x=='Petrol':
        return 5
```

```python
new_df['Tran_Fuel']=new_df['Fuel'].apply(trans_Fuel)
new_df
```

# Apply One Hot Encode on Brand Column And Merged with Data set

```python
def trans_Brand(x):
    if x=='Suzuki':
        return 1
    if x=='Toyota':
        return 2
    if x=='Honda':
        return 3
    if x=='Daihatsu':
        return 4
    if x=='Nissan':
        return 5
    if x=='Mitsubishi':
        return 6
    if x=='Hyundai':
        return 7
    if x=='Other Brands':
        return 8
    if x=='KIA':
        return 9
    if x=='Mazda':
        return 10
    if x=='Mercedes':
        return 11
    if x=='FAW':
        return 12
    if x=='Daewoo':
        return 13
    if x=='Chevrolet':
        return 14
    if x=='BMW':
        return 15
    if x=='Lexus':
        return 16
    if x=='Subaru':
        return 17
    if x=='Audi':
        return 18
    if x=='Classic & Antiques':
        return 19
    if x=='Changan':
        return 20
    if x=='Range Rover':
        return 21
    if x=='Porsche':
        return 22
    if x=='Land Rover':
        return 23
```

```python
new_df['Tran_Brand']=new_df['Brand'].apply(trans_Brand)
new_df
```

```python
def gradient_descent(x,y):
    m_curr=b_curr=0
    iterations=50
    n=len(x)
    learning_rate=0.001
        
    for i in range(iterations):
        y_predicted=m_curr*x + b_curr
        cost=(1/n) * sum([val**2 for val in (y-y_predicted)])
        md= -(2/n)*sum(x*(y-y_predicted))
        bd= -(2/n)*sum(y-y_predicted)
        m_curr=m_curr - learning_rate * md
        b_curr=b_curr - learning_rate * bd
        print("m = {},b = {} ,Cost = {},iteration = {}".format(m_curr,b_curr,cost,i))
```

```python
x=np.array([1,2,3,4,5])
y=np.array([5,7,9,11,15])
gradient_descent(x,y)
```
